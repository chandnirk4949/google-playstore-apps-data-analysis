# arranging columns to make it more viable : changing datatypes,renaming etc

```python
#To get integer columns, float columns,object columns,boolean columns,datetimecolumns
int_cols = df1.dtypes[df1.dtypes=='int64'].index.to_list()
float_cols = df1.select_dtypes(include=['float64']).columns.to_list()
obj_cols = df1.select_dtypes(include=['object']).columns.to_list()
bool_cols = df1.select_dtypes(include=['boolean']).columns.to_list()
dt_cols = df1.select_dtypes(include=['datetime64']).columns.to_list()
```

```python
print('total columns :    ', df1.columns.to_list(),'\n')
print('integer columns:   ', int_cols, '\n')
print('float columns:     ', float_cols, '\n')
print('object columns:    ', obj_cols,  '\n')
print('boolean columns:   ', bool_cols,  '\n')
print('datetime columns:  ', dt_cols,  '\n')
```

```python
#to change datetime and boolean formats
df1['Released']=df1['Released'].astype('datetime64')
df1['Last Updated']=df1['Last Updated'].astype('datetime64')

df1['Free']=df1['Free'].astype('int64')
df1['Ad Supported']=df1['Ad Supported'].astype('int64')
df1['In App Purchases']=df1['In App Purchases'].astype('int64')
df1['Editors Choice']=df1['Editors Choice'].astype('int64')
```
